```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
```

```python
url = "https://www.trc-canada.com/products-listing"
client = urlopen(url)
html = client.read()
client.close()
pg_soup = soup(html, 'html.parser')
```

```python
containers = pg_soup.findAll('div',{'class': 'ContentDesc'})
# print(containers[:])
```

```python
b='\033[1m'
be = '\033[0m'

# Get the title
title = pg_soup.find('title').contents
print(b,title[0].upper(),be)

# ////******  ----------------------------------- Web Scrapping Program    ---------------------------  ******** //////
#     ////-------------------------------------- For First 30 Element Groups --------------------------------/////


label =0
for i in range(0,180,6):
    label+=1
    print('Index',label)
    # Get Catalogue Num
    
    raw_label = containers[i].find_all({'a':'href'})
    raw_label = str(raw_label)
    if(containers[i].find('label')!=None):
        cat_label = containers[i].find('label').text
        if(cat_label=='Catalogue No: '):    
            if(len(raw_label)>20):
                cat_num = raw_label[34:41]
                print(b,cat_label,be,cat_num)
            else:
                cat_num1 = containers[i].contents
                cat_num1 = ''.join(cat_num1[2].split())
                print(b,cat_label,be,cat_num1)
        else:
            pass
    else:
    
        pass
    
    
    i+=1
    #Get Chemical Name Title
    chem_name = containers[i].contents
    if(containers[i].find('label')!=None):
        chem_label = containers[i].find('label').text
        if(chem_label=='Chemical Name: '):
            print(b,chem_label,be,end="")
            if(chem_name!=chem_label): 
                print(''.join(chem_name[2].split()))
            else:
                print('---') 
        else:
            pass
    else:
        pass
    
    i+=1
    # Get CAS Number
    cas_name = containers[i].contents
    if(containers[i].find('label')!=None):
        cas_label = containers[i].find('label').text
        if(cas_label=='CAS number: '):
            print(b,cas_label,be,end="")
            if(cas_name!=cas_label):
                print(''.join(cas_name[2].split()))        
            else:
                print('---') 
        else:
            pass
    else:
        pass    
    
    i+=1
    # For Alternative CAS name
    Alcas_name = containers[i].contents
    if(containers[i].find('label')!=None):
        Alcas_label = containers[i].find('label').text
        if(Alcas_label=='Alternate CAS: '):
            print(Alcas_label, end="")
            if(Alcas_name!=Alcas_label):
                print(''.join(Alcas_name[2].split()))
        
            else:
                print('---') 
        else:
            pass
    else:
        pass   

    
    i+=1
    #obtain molecular Formula
    mol_name = containers[i].contents
    if(containers[i].find('label')!=None):
        mol_label = containers[i].find('label').text
        if(mol_label=='Mol. Formula: '):
            print(mol_label, end="")
            if(mol_name!=mol_label):
                print(''.join(mol_name[2].split()))
        
            else:
                print('---') 
        else:
            pass
    else:
        pass 

    i+=1
    #obtain Synonyms
    syn_name = containers[i].contents
    if(containers[i].find('label')!=None):
        syn_label = containers[i].find('label').text
        if(syn_label=='Synonyms: '):
            print(syn_label, end="")
            if(syn_name!=syn_label):
                print(''.join(syn_name[2].split()))
        
            else:
                print('---') 
        else:
            pass
    else:
        pass 
    
    
        
    

    print('\n\n')
```
