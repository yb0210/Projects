```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
cwd = os.chdir("C:/Users/Amit/Desktop/New folder/Kaleidofin/Assignment_1 2")

file = 'chit fund exercise.xlsx'
df = pd.read_excel(file,sheet_name='Chit fund cash flows')
# pd.DataFrame(df)
df.head()
```

```python
df.plot()
df.plot(subplots=True)
plt.show()
```

## Annualized Return for first month

```python
#compound annual growth rate is 

#beginning value = bv; Ending Value =ev ; Timeperiod = t;
t = 1/12 #Time Period

#for first month Annual return for non winner
print('For the First month investment the annual return of a non winner would be:', end=' ')
f_bv = df.values[0,1] #beginning value in first month
f_ev = df.values[0,5] #End Value in first month

f_cagr = (((1+(f_ev-f_bv)/f_bv)**(1/t))-1)*100
# f_cagr = '{0:.2f}'.format(round(f_cagr,1))
print (f_cagr,'%')

#for first month Annual return for winner
print('For the First month investment the annual return of a winner would be:', end=' ')
f_wbv = df.values[0,1] #Winnwer's Beginning Value in first month
f_wev = df.values[0,4] #Winnwer's End Value in first month

f_wcagr = (((1+(f_wev-f_wbv)/f_wbv)**(1/t))-1)*100
# f_wcagr = '{0:.2f}'.format(round(f_wcagr,1))
print (f_wcagr,'%')
```

```python

plt.title('First month investment the annual return')
x = ['Investment' ,'Non_w Return']
y = [f_bv,f_ev]
x1=['Investment' ,'w_Return']
y1 = [f_wbv,f_wev]

plt.xlabel('Participant')
plt. ylabel('Amount')
plt.bar(x,y,label = 'Non-winner',color = 'r')
plt.bar(x1,y1,label = 'Winner',color = 'g')
plt.legend()
plt.show()
```

## Annualized Return for last month

## Return % of Each monthbid winner

```python
t = 25/12

#for Annual return for winner till the chit fund lasts
print('For the Over all investment return of a winner would be:')
for i in range(len(df.values)):
    wbv = df.values[i,1]
    wev = df.values[i,4]
    
    wcagr = (((1+(wev-wbv)/wbv)**(1/t))-1)*100
    wcagr =round(wcagr,2)
    print (wcagr,'%')
```
