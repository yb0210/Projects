```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
cwd = os.chdir("D:/Python Software/project/Kaleidofin/Chitfund/Assignment_1 2")

file = 'chit fund exercise.xlsx'
df = pd.read_excel(file,sheet_name='Chit fund cash flows')
# pd.DataFrame(df)
df.head()
```

```python
df.plot()
df.plot(subplots=True)
plt.show()
```

## Annualized Return for first month

```python
#compound annual growth rate is 

#beginning value = bv; Ending Value =ev ; Timeperiod = t;
t = 1/12 #Time Period

#for first month Annual return for non winner
print('For the First month investment the annual return of a non winner would be:', end=' ')
f_bv = df.values[0,1] #beginning value in first month
f_ev = df.values[0,5] #End Value in first month

f_ret = (f_ev-f_bv)/f_bv
f_cagr = (((1+f_ret)**(1/t))-1)*100
# f_cagr = '{0:.2f}'.format(round(f_cagr,1))
print (f_cagr,'%')

#for first month Annual return for winner
print('For the First month investment the annual return of a winner would be:', end=' ')
f_wbv = df.values[0,1] #Winnwer's Beginning Value in first month
f_wev = df.values[0,4] #Winnwer's End Value in first month

f_wret = (f_wev-f_wbv)/f_wbv
f_wcagr = (((1+f_wret)**(1/t))-1)*100
# f_wcagr = '{0:.2f}'.format(round(f_wcagr,1))
print (f_wcagr,'%')
```

```python
plt.title('First month investment the annual return')
x = ['Investment']
y = [m for m in range(0,f_bv)] #investmnet are same irrespective of results
x1 = ['Non_w Return']
y1 = [ev for ev in range(0,f_ev)] #To plot the bar of Non_winner
x2 =['w_Return']
y2 = ['Investment']
y2 = [wev for wev in range(0,f_wev)] #To plot the bar of bid_winner



plt.xlabel('Participant')
plt. ylabel('Amount')
plt.bar(x,y,label = 'Investment',color = 'c')
plt.bar(x1,y1,label = 'Non-winner',color = 'r')
plt.bar(x2,y2,label = 'Winner',color = 'g')

plt.legend()
plt.show()
```

## Annualized Return for last month

```python
#compound annual growth rate is 

#beginning value = bv; Ending Value =ev ; Timeperiod = t;
t = 1/12 #Time Period

#for first month Annual return for non winner
print('For the last month investment the annual return of a non winner would be:', end=' ')
l_bv = df.values[24,1] #beginning value in last month
l_ev = df.values[24,5] #End Value in last month

l_ret = (l_ev-l_bv)/l_bv
l_cagr = (((1+l_ret)**(1/t))-1)*100
# l_cagr = '{0:.2f}'.format(round(l_cagr,1))
print (l_cagr,'%')

#for first month Annual return for winner
print('For the Last month investment the annual return of a winner would be:', end=' ')
l_wbv = df.values[24,1]
l_wev = df.values[24,4]

l_wret = (l_wev-l_wbv)/l_wbv
l_wcagr = (((1+l_wret)**(1/t))-1)*100
# l_wcagr ='{0:.2f}'.format(round(l_wcagr,2))
print (l_wcagr,'%')
```

```python
plt.title('Last month investment the annual return')

x = ['Investment']
y = [m for m in range(0,f_bv)] #investmnet are same irrespective of results
x1 = ['Investment' ,'Non_w Return']
y1 = [l_bv,l_ev]
x2=['Investment' ,'w_Return']
y2 = [l_wbv,l_wev]

plt.xlabel('Participant')
plt. ylabel('Amount')
plt.bar(x1,y1,label = 'Non-winner',color = 'r')
plt.bar(x2,y2,label = 'Winner',color = 'g')
plt.bar(x,y,label = 'Investment',color = 'c')

plt.legend()
plt.show()
```

## Return % of Each monthbid winner

```python
t = 1/12

#for Annual return for winner till the chit fund lasts
print('For the Over all investment return of a winner would be:')
for i in range(len(df.values)):
    wbv = df.values[i,1]
    wev = df.values[i,4]
    
    wret = (wev-wbv)/wbv
    wcagr = (((1+wret)**(1/t))-1)*100
    print (wcagr,'%')
```

```python
plt.title('Winners Overall Return for 25 months')
plt.xlabel('No. Of Months')
plt.ylabel('Percentage hike in Return (in exponents)')
for j in range(len(df.values)):
    wbv = df.values[j,1]
    wev = df.values[j,4]
    wcagr = (((1+(wev-wbv)/wbv)**(1/t))-1)*100
    ++j
    plt.scatter(j,[wcagr], color='b', edgecolors='r')
plt.grid(True)
plt.show
    
```
